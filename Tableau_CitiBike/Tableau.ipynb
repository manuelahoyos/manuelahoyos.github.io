```python
import pandas as pd
import numpy as np
import datetime
import calendar
```

```python
January = pd.read_csv(r'201901.csv', error_bad_lines=False)
```

```python
January['Month'] = pd.DatetimeIndex(January['starttime']).month
January['Month'] = January['Month'].apply(lambda x: calendar.month_abbr[x])
January['Age'] = 2019 - January['birth year']
January
```

```python
JanuaryRenamed = January.rename(columns={"tripduration": "Trip Duration (seconds)", "usertype": "User Type",
                                        "gender":"Gender"})
JanuaryRenamed
```

```python
JanuaryFinal = JanuaryRenamed.drop(["starttime","stoptime","start station id", "start station name", "start station latitude",
                             "start station longitude","end station id", "end station name", "end station latitude",
                            "end station longitude", "bikeid", "birth year"], axis=1)
JanuaryFinal['Gender'].replace([0,1,2],['Unknown','Male','Female'],inplace=True)
JanuaryFinal.head()
```

```python
June = pd.read_csv(r'201906.csv', error_bad_lines=False)
June['Month'] = pd.DatetimeIndex(June['starttime']).month
June['Month'] = June['Month'].apply(lambda x: calendar.month_abbr[x])
June['Age'] = 2019 - June['birth year']
JuneRenamed = June.rename(columns={"tripduration": "Trip Duration (seconds)", "usertype": "User Type",
                                        "gender":"Gender"})
JuneFinal = JuneRenamed.drop(["starttime","stoptime","start station id", "start station name", "start station latitude",
                             "start station longitude","end station id", "end station name", "end station latitude",
                            "end station longitude", "bikeid", "birth year"], axis=1)
JuneFinal['Gender'].replace([0,1,2],['Unknown','Male','Female'],inplace=True)
JuneFinal.head()
```

```python
October = pd.read_csv(r'201910.csv', error_bad_lines=False)
October['Month'] = pd.DatetimeIndex(October['starttime']).month
October['Month'] = October['Month'].apply(lambda x: calendar.month_abbr[x])
October['Age'] = 2019 - October['birth year']
OctoberRenamed = October.rename(columns={"tripduration": "Trip Duration (seconds)", "usertype": "User Type",
                                        "gender":"Gender"})
OctoberFinal = OctoberRenamed.drop(["starttime","stoptime","start station id", "start station name", "start station latitude",
                             "start station longitude","end station id", "end station name", "end station latitude",
                            "end station longitude", "bikeid", "birth year"], axis=1)
OctoberFinal['Gender'].replace([0,1,2],['Unknown','Male','Female'],inplace=True)
OctoberFinal.head()
```

```python
all_dfs = [JanuaryFinal, JuneFinal, OctoberFinal]
Final = pd.concat(all_dfs).reset_index(drop=True)
Final
```

```python
Final.to_csv(r'Final.csv')
```
